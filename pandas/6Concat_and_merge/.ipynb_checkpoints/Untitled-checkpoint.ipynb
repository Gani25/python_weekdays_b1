```python
import pandas as pd
```

```python
cricket_player_india = pd.DataFrame({
    "player_name":["Virat Kohli","Sachin Tendulkar","Rohit Sharma"],
    "runs":[100,150,90],
    "player_type":["Batsman","Spinner","Bowler"]
}
)
cricket_player_india
```

```python
cricket_player_australia = pd.DataFrame({
    "player_name":["David Warner","Ryan Rickelton","ABD Villers"],
    "runs":[85,120,190],
    "player_type":["Bowler","Batsman","Wicket Keeper"]
}
)
cricket_player_australia
```

```python
pd.concat([cricket_player_india, cricket_player_australia])
```

```python
pd.concat([cricket_player_india, cricket_player_australia], ignore_index=True)
```

```python
players = pd.concat([cricket_player_india, cricket_player_australia],keys=["India Player","Australia Player"])
players
```

```python
players.loc["Australia Player"]
```

```python
players.loc["Australia Player"].iloc[0]
```

# Merge

```python
students = pd.DataFrame({
    "roll_no": [1,2,3,4,5,6,7],
    "name":["d1","d2","d3","d4","d5","d6","d7"],
    "std":[12,15,"BE","dip",10,16,14]
})
students
```

```python
courses = pd.DataFrame({
    "cid":[1001,1002,1001,1003,1002],
    "course_name":["Python","Java","Python","Power BI","Java"],
    "roll_no":[2,7,1,2,7]
})
courses
```

# Inner Join

```python
pd.merge(students,courses, on="roll_no")
```

# Left

```python
pd.merge(students,courses, on="roll_no", how="left")
```

```python
pd.merge(students,courses, on="roll_no", how="right")
```

```python
pd.merge(students,courses, on="roll_no", how="outer")
```

```python
pd.merge(students,courses, how="cross")
```
