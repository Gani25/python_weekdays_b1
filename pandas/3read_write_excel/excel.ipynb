```python
df_student = pd.read_excel("students.xlsx", skiprows=1, na_values=["N.A","NOT AVAILABLE"])
df_student
```

```python
df_student = pd.read_excel("students.xlsx", skiprows=1, na_values=["N.A","NOT AVAILABLE","n.a"])
df_student
```

```python
df_student = pd.read_excel("students.xlsx", skiprows=1, 
                           na_values={
                               "Gender":["N.A","n.a"], 
                               "Phone":["n.a","NOT AVAILABLE"]
                           }, names=["student_id", "sname","gender","phone"])
df_student
```
