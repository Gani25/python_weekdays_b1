```python
df_marks = pd.read_excel("students.xlsx","Sheet2", header=1, na_values=[-1])
df_marks
```

```python
# Forcefull conversion
import numpy as np
```

```python

df_marks.replace(-1, np.nan, inplace=True)
df_marks
```

```python
df_student
```

```python
student_merged = pd.merge(df_student, df_marks, left_on="student_id",right_on="rollno")
student_merged
```

```python
student_merged.drop("rollno", inplace=True, axis=1)
student_merged
```

```python
student_merged["percentage"] = ((student_merged.english + student_merged.maths+ student_merged.science+student_merged.hindi+student_merged.marathi) / student_merged.total) * 100

student_merged
```

```python
student_merged.to_excel( excel_writer="students_merged.xlsx",sheet_name="merged",index=False )
```

```python
student_merged.to_csv("students_merged.csv", index=False)
```
