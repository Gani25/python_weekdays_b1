# Find Max, Min and Average Ratings For All Movies

```python
all_movies.imdb_rating.head()
```

```python
print(f"Max Rating of All movies {all_movies.imdb_rating.max()}")
print(f"Min Rating of All movies {all_movies.imdb_rating.min()}")
print(f"Average Rating of All movies {all_movies.imdb_rating.mean().round(2)}")
```

```python
bolly_movies = all_movies[all_movies.industry == "Bollywood"]
bolly_movies
```

```python
print(f"Max Rating of Bollywood movies {bolly_movies.imdb_rating.max()}")
print(f"Min Rating of Bollywood movies {bolly_movies.imdb_rating.min()}")
print(f"Average Rating of Bollywood movies {bolly_movies.imdb_rating.mean().round(2)}")
```
