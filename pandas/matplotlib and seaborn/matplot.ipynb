```python
import matplotlib.pyplot as plt
```

```python
sales = pd.DataFrame(
    {
        "Quarter":["Q1 2024","Q2 2024","Q3 2024","Q4 2024","Q1 2025","Q2 2025","Q3 2025"],
        "Iphone":[50,40,33,65,52,20,39],
        "Samsung Fold":[10,45,25,20,15,12,25],
        "Redmi":[80,30,105,115,60,80,70]
    }
)
sales
```

```python
plt.figure(figsize=(12,6))
plt.plot(sales["Quarter"],sales["Iphone"], marker = "o", label="Iphone", color="red", )

plt.plot(sales["Quarter"],sales["Samsung Fold"], marker = "p", label="Samsung Fold", color="black")

plt.plot(sales["Quarter"],sales["Redmi"], marker = "^", label="Redmi", color="Green")
plt.grid(linestyle = "--", alpha = 0.3)

plt.legend(title = "Products")
plt.xlabel("Products", fontsize = 20)
plt.ylabel("Quantity in K", fontsize = 20)
plt.title("Quantity Of Products Sold In Quarter", fontsize = 20, pad=20)
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.plot(sales["Quarter"],sales["Iphone"], marker = "o", label="Iphone", color="crimson", 
         linewidth = 2.5, markersize = 10, mfc='black', mec='blue')

plt.plot(sales["Quarter"],sales["Samsung Fold"], marker = "p", label="Samsung Fold", color="black", linewidth = 2.5, markersize = 10)

plt.plot(sales["Quarter"],sales["Redmi"], marker = "^", label="Redmi", color="Green", linewidth = 2.5, markersize = 10)
plt.grid(linestyle = "--", alpha = 0.3)
plt.title("Quantity Of Products Sold In Quarter", fontsize = 20, pad=20)
plt.legend(title = "Products", fontsize = 14, title_fontsize = 14, loc = "upper left", shadow=True, frameon=True)
plt.xlabel("Products", fontsize = 20, labelpad=20)
plt.ylabel("Quantity in K", fontsize = 20, labelpad=20)
custom_yticks = range(0,121,10)
plt.yticks(ticks=custom_yticks, fontsize = 14)
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

# Pie Chart

```python
total_sales = sales[["Iphone","Samsung Fold","Redmi"]].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.figure(figsize=(7,7))
plt.pie(
    total_sales, 
    labels=total_sales.index,
    startangle=0,
    autopct="%.2f%%",
    colors=["skyblue","orange","lightgreen"],
    wedgeprops = {"edgecolor":"white"},
    shadow=True,
    labeldistance = 0.5,
    pctdistance = 1.1,
    explode = (0.3,0.1,0)
)
plt.title("Total Sales Of Products", fontsize = 20, pad= 20)
plt.show()
```
