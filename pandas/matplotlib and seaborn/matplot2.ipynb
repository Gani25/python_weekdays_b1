# Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scores = pd.read_csv("scores.csv")
scores.head()
```

```python
plt.hist(scores["Marks"])
plt.show()
```

```python
import seaborn as sns
```

```python
sns.histplot(scores.Marks, kde=True)

xvalues = range(0,101,10)
plt.xticks(ticks = xvalues)
plt.xlabel(xlabel="Marks", fontsize = 16, labelpad=16)
plt.ylabel(ylabel="Counts", fontsize = 16, labelpad=16)
plt.show()
```
