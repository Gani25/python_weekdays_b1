# Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
scores = pd.read_csv("scores.csv")
scores.head()
```

```python
plt.hist(scores["Marks"])
plt.show()
```

```python
import seaborn as sns
```

```python
sns.histplot(scores.Marks, kde=True)

xvalues = range(0,101,10)

plt.xticks(ticks = xvalues)
plt.xlabel(xlabel="Marks", fontsize = 16, labelpad=16)
plt.ylabel(ylabel="Counts", fontsize = 16, labelpad=16)
plt.show()
```

```python

sns.set_style("whitegrid")
hist = sns.histplot(scores.Marks, 
             kde=True, 
             bins = 10, 
             color="red", 
             alpha=0.7, 
             edgecolor = "black"
            )

# display count on each path
for patch in hist.patches:
    height = patch.get_height()
    if height > 0:
        hist.text(patch.get_x() + patch.get_width()/2,
        height + 0.5,
                  f"{int(height)}",
                  ha="center",
                  fontsize = 10,
                  fontweight = "bold"
                 )

plt.xticks(ticks = np.arange(0,101,10), fontsize=12)
plt.xlabel(xlabel="Marks", fontsize = 12, labelpad=16, fontweight="bold")
plt.ylabel(ylabel="Counts", fontsize = 12, labelpad=16, fontweight="bold")
plt.title("Marks Distribution", fontsize = 14, fontweight ="bold",pad=16, color="DarkBlue"  )
plt.show()

```
